```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import requests
import bs4
```

```python
request = requests.get('https://en.wikipedia.org/wiki/Pok%C3%A9mon')
```

```python
soup = bs4.BeautifulSoup(request.text, "lxml")
```

```python
from htmllaundry import sanitize, strip_markup
```

```python
bad_html = sanitize(soup)
```

```python
text = strip_markup(bad_html)
```

```python
cloud = WordCloud(width=3000, height= 2000, random_state=1,background_color= , colormap= "Set1", stopwords = STOPWORDS).generate(text)
plt.figure(figsize=(40,30), facecolor=None)
plt.imshow(cloud, interpolation= "bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
cloud.to_file('Pokemon Wikipedia Page WC v4.png')
```
